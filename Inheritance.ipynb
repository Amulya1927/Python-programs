```python
class Parent:
    def func1(self):
       print("This function is in parent class") 
        
class Child (Parent):
     def func2(self):
        print("This function is in child class")
object=Child()
object.func1()
object.func2()
```

```python
class Grandfather:
    def land(self):
        print("Grandfather has land")
class Father(Grandfather):
    def bike(self):
        print("father has a bike")
class Son(Father):
    def mobile(self):
        print("son has a mobile")
s =  Son()
s.land()
s.bike()
s.mobile()
```
